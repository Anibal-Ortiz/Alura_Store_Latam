### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación

```python
suma_total_tienda1 = tienda['Precio'].sum()
suma_total_tienda2 = tienda2['Precio'].sum()
suma_total_tienda3 = tienda3['Precio'].sum()
suma_total_tienda4 = tienda4['Precio'].sum()

print("Suma total tienda 1:", suma_total_tienda1)
print("Suma total tienda 2:", suma_total_tienda2)
print("Suma total tienda 3:", suma_total_tienda3)
print("Suma total tienda 4:", suma_total_tienda4)

import matplotlib.pyplot as plt

# 1. Preparar los datos para el gráfico
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores_ingresos = [suma_total_tienda1, suma_total_tienda2, suma_total_tienda3, suma_total_tienda4]

# 2. Crear el gráfico
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, valores_ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# 3. Añadir títulos y etiquetas
plt.title('Ingresos Totales por Tienda', fontsize=14)
plt.ylabel('Ingresos (en Miles de Millones $)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# --- Análisis de Valoración Media por Tienda (Versión directa sin gráficos) ---

# 1. Define el nombre de la columna
columna_valoracion = 'Calificación'

# 2. Calcular la media (promedio) para cada tienda
media_tienda1 = tienda[columna_valoracion].mean()
media_tienda2 = tienda2[columna_valoracion].mean()
media_tienda3 = tienda3[columna_valoracion].mean()
media_tienda4 = tienda4[columna_valoracion].mean()

# 3. Imprimir los resultados formateados
print("--- Iniciando Análisis de Valoración Media por Tienda ---")
print(f"La valoración media para la Tienda 1 es: {media_tienda1:.2f} estrellas")
print(f"La valoración media para la Tienda 2 es: {media_tienda2:.2f} estrellas")
print(f"La valoración media para la Tienda 3 es: {media_tienda3:.2f} estrellas")
print(f"La valoración media para la Tienda 4 es: {media_tienda4:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
# --- Análisis de Productos Más y Menos Vendidos (Versión directa sin gráficos) ---

columna_producto = 'Producto'

print("--- Iniciando Análisis de Productos ---")

# --- Tienda 1 ---
conteo_1 = tienda[columna_producto].value_counts()
print("\n--- Tienda 1: MÁS Vendidos (Top 10) ---")
print(conteo_1.head(10))
print("\n--- Tienda 1: MENOS Vendidos (Bottom 10) ---")
print(conteo_1.tail(10).sort_values(ascending=True))

# --- Tienda 2 ---
conteo_2 = tienda2[columna_producto].value_counts()
print("\n\n--- Tienda 2: MÁS Vendidos (Top 10) ---")
print(conteo_2.head(10))
print("\n--- Tienda 2: MENOS Vendidos (Bottom 10) ---")
print(conteo_2.tail(10).sort_values(ascending=True))

# --- Tienda 3 ---
conteo_3 = tienda3[columna_producto].value_counts()
print("\n\n--- Tienda 3: MÁS Vendidos (Top 10) ---")
print(conteo_3.head(10))
print("\n--- Tienda 3: MENOS Vendidos (Bottom 10) ---")
print(conteo_3.tail(10).sort_values(ascending=True))

# --- Tienda 4 ---
conteo_4 = tienda4[columna_producto].value_counts()
print("\n\n--- Tienda 4: MÁS Vendidos (Top 10) ---")
print(conteo_4.head(10))
print("\n--- Tienda 4: MENOS Vendidos (Bottom 10) ---")
print(conteo_4.tail(10).sort_values(ascending=True))
```

2. Desarrollo y Análisis de Datos
Se analizaron los datos de las cuatro tiendas. Si bien superficialmente parecen similares, un análisis detallado de las métricas de eficiencia y satisfacción del cliente revela diferencias clave.

B. Análisis de Productos y Categorías
El análisis de categorías (Gráfico 2: Ventas por Categoría) y de productos individuales (Paso 4) muestra que las cuatro tiendas tienen un comportamiento de mercado muy similar. Vender la Tienda 1 le permite eliminar su operación más problemática y costosa, maximizando al mismo tiempo el retorno de su inversión.
